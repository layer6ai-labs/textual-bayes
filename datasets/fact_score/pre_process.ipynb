```python
data = []
with open('fact_score_names.txt', 'r') as file:
    for count, line in enumerate(file, start=1):
        name = line.strip()
        data.append({"question": f"Tell me a bio about {name}", "id": count})

print(len(data))
```

```python
import random

# Randomly sample 200 items for training
random.seed(42)
train_data = random.sample(data, 200)

# Use set difference to get the remaining 300 items for testing
test_data = [item for item in data if item not in train_data]

print(f"Train size: {len(train_data)}, Test size: {len(test_data)}")
```

```python
import json

# Write train_data to a JSONL file
with open('factscore_train.jsonl', 'w') as train_file:
    for item in train_data:
        json.dump(item, train_file, ensure_ascii=False)
        train_file.write('\n')

with open('factscore_test_additional.jsonl', 'w') as test_file:
    for item in test_data:
        json.dump(item, test_file, ensure_ascii=False)
        test_file.write('\n')

```
